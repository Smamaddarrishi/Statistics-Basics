```python
#Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers. numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

import statistics
from collections import Counter

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Compute Mean
mean_value = statistics.mean(numbers)

# Compute Median
median_value = statistics.median(numbers)

# Compute Mode(s) - using Counter for accurate multi-modal result
counts = Counter(numbers)
max_count = max(counts.values())
modes = [key for key, value in counts.items() if value == max_count]

print(f"List of numbers: {numbers}")
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode(s): {modes}")
```

```python
#Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python: list_x=[10,20,30,40,50] list_y=[15,25,35,45,60]

import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

x = np.array(list_x)
y = np.array(list_y)

# 1. Compute Covariance
# np.cov returns a covariance matrix; [0, 1] element is Cov(X, Y)
covariance_matrix = np.cov(x, y)
covariance = covariance_matrix[0, 1]

# 2. Compute Correlation Coefficient (Pearson's r)
# np.corrcoef returns a correlation matrix; [0, 1] element is Corr(X, Y)
correlation_matrix = np.corrcoef(x, y)
correlation_coefficient = correlation_matrix[0, 1]

print(f"list_x: {list_x}")
print(f"list_y: {list_y}")
print(f"Covariance (Cov(X, Y)): {covariance:.2f}")
print(f"Correlation Coefficient (Corr(X, Y)): {correlation_coefficient:.4f}")
```

```python
#Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result: data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Calculate Quartiles and IQR to define fences
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

# Identify Outliers
outliers = [x for x in data if x < lower_fence or x > upper_fence]

# Create Boxplot
plt.figure(figsize=(7, 4))
plt.boxplot(data, vert=False)
plt.title('Boxplot of Numeric Data')
plt.xlabel('Data Values')
# plt.show() # To display the plot

print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR: {IQR}")
print(f"Upper Fence: {upper_fence}")
print(f"Identified Outliers: {outliers}")
```

```python
#Question 8: Write Python code to compute the correlation between the two lists: advertising_spend = [200, 250, 300, 400, 500] daily_sales = [2200, 2450, 2750, 3200, 4000]

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

ad_spend = np.array(advertising_spend)
sales = np.array(daily_sales)

# Compute Correlation Coefficient
correlation_matrix = np.corrcoef(ad_spend, sales)
correlation_coefficient = correlation_matrix[0, 1]

print(f"Advertising Spend: {advertising_spend}")
print(f"Daily Sales: {daily_sales}")
print(f"Correlation Coefficient: {correlation_coefficient:.4f}")
```

```python
#Question 9: Write Python code to create a histogram using Matplotlib for the survey data: survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Create Histogram
plt.figure(figsize=(8, 5))
# Set bins for clarity, aligning bars with integer scores 4 through 10
min_score = min(survey_scores)
max_score = max(survey_scores)
bins = np.arange(min_score, max_score + 2) - 0.5

plt.hist(survey_scores, bins=bins, edgecolor='black', alpha=0.7)
plt.xticks(range(min_score, max_score + 1))
plt.title('Histogram of Customer Satisfaction Survey Scores (1-10 Scale)')
plt.xlabel('Satisfaction Score')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.5)
# plt.show() # To display the plot

print(f"Survey Scores: {survey_scores}")
print("\n--- Histogram Generated ---")
# Print frequency counts for data analysis:
from collections import Counter
score_counts = Counter(survey_scores)
print(f"Score Frequencies: {score_counts}")
```
